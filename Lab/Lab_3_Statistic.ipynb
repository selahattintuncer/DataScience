```python
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
from scipy import stats
```

```python
tv_hours = [3,5,2,0,5,3,1,4,3,4] # öğrencilerin tv izleme süresi
GPA = [2.7,2.1,3.3,3.4,2.0,3.0,3.6,2.8,3.5,2.6] # not ortalaması
```

```python
plt.scatter(tv_hours, GPA);
```

```python
plt.figure(figsize=(8,6))
plt.figure(figsize=(8,6))
ax = plt.axes()
ax.scatter(tv_hours, GPA)
ax.set_xlabel("TV Hours")
ax.set_ylabel("GPA")
ax.axis("tight")
plt.show()
```

```python
x = np.array(tv_hours)
y = np.array(GPA)
```

```python
len(x)*sum(x*y) - sum(x) * sum(y)
```

```python
np.sqrt((len(x)*sum(x**2)-sum(x)**2)*(len(y)*sum(y**2)-sum(y)**2))
```

```python
-71/82.84926
```

```python
stats.pearsonr(tv_hours, GPA)  # ilk sayı pearson korelasyon katsayısı, ikincisi ise p değeri
```

```python
np.corrcoef(tv_hours,GPA)  # bu da korelasyonu verir.
```

```python
stats.spearmanr(GPA, tv_hours)  # spearman korelasyonu veriri
```

```python
stats.kendalltau(GPA, tv_hours) # kendal korelasyonu verir
```

```python
np.cov(GPA, tv_hours)  # covariance matrix verir
```

```python
stats.tvar(GPA)  # sample varyans
```

```python
stats.tvar(tv_hours)   # sample varyans
```

```python
# numpy ile korelasyon katsayısı

np.corrcoef(tv_hours, GPA)  # buradaki 1 kendisi ile olan korelasyon, DEfault olarak pearson korelasyonunu verir.
```
